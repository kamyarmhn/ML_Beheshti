# تمرین 7.3 (PRML): کافی‌بودن دو نقطهٔ داده (یکی از هر کلاس) برای تعیین مکان ابرصفحهٔ بیشینه‌حاشیه

هدف: نشان دادن این‌که صرف‌نظر از بُعدِ فضا، داشتن فقط دو نقطهٔ داده—یکی با برچسب +1 و دیگری با برچسب −1—برای تعیین مکان ابرصفحهٔ بیشینه‌حاشیه کافی است.


## ۱) صورت‌بندی استاندارد مسئله (دادهٔ کاملاً جداشدنی)

می‌خواهیم ابرصفحهٔ جداساز را به‌صورت \(w^T x + b = 0\) بیابیم که حاشیه را بیشینه می‌کند. این کار با کمینه‌سازی نرمِ \(w\) تحت قیود حاشیه انجام می‌شود.



$$
\min_{w,\,b}\ \tfrac{1}{2}\,\|w\|^2
$$

$$
\text{s.t.}\quad y_i\,(w^T x_i + b) \ge 1,\quad i=1,2,\ \ y_1=+1,\ y_2=-1
$$


این قیود برای دو نمونه به شکل زیر هستند.



$$
w^T x_1 + b \ \ge\ 1
$$

$$
-(w^T x_2 + b) \ \ge\ 1 \ \ \Longleftrightarrow\ \ w^T x_2 + b \ \le\ -1
$$


## ۲) استدلال هندسی و تبدیل قیود به تساوی (نقاط پشتیبان)

برای بیشینه‌شدن حاشیه، هر دو نمونه باید روی مرزهای حاشیه قرار گیرند (هر دو **بردار پشتیبان**). در نتیجه قیود به **تساوی** تبدیل می‌شوند.



$$
w^T x_1 + b \ =\ 1
$$

$$
w^T x_2 + b \ =\ -1
$$


اکنون یک دستگاه دو معادله‌ای در دو مجهول (\(w\) و \(b\)) داریم. با تفریق این دو معادله، \(b\) حذف می‌شود.



$$
(w^T x_1 + b) - (w^T x_2 + b) \ =\ 1 - (-1)
$$

$$
w^T (x_1 - x_2) \ =\ 2
$$


## ۳) هم‌راستایی \(w\) با خطِ واصل و محاسبهٔ بسته

از دید هندسی، \(w\) باید با بردارِ واصلِ دو نقطه هم‌راستا باشد، پس می‌نویسیم \(w = \,\alpha\,(x_1 - x_2)\). با جایگذاری در رابطهٔ بالا، \(\alpha\) را به‌دست می‌آوریم.



$$
(\alpha (x_1 - x_2))^T (x_1 - x_2) \ =\ 2
$$

$$
\alpha\,\|x_1 - x_2\|^2 \ =\ 2 \ \ \Longrightarrow\ \ \alpha \ =\ \frac{2}{\|x_1 - x_2\|^2}
$$

$$
w \ =\ \frac{2}{\|x_1 - x_2\|^2}\,(x_1 - x_2)
$$


برای \(b\)، دو تساویِ مرزی را با هم جمع می‌کنیم تا به فرمول بسته برسیم.



$$
(w^T x_1 + b) + (w^T x_2 + b) \ =\ 1 + (-1) \ =\ 0
$$

$$
w^T (x_1 + x_2) + 2b \ =\ 0 \ \ \Longrightarrow\ \ b \ =\ -\tfrac{1}{2}\, w^T (x_1 + x_2)
$$

$$
b \ =\ -\tfrac{1}{2}\,\Big(\tfrac{2}{\|x_1 - x_2\|^2} (x_1 - x_2)\Big)^T (x_1 + x_2)
\ =\ -\,\frac{(x_1 - x_2)^T(x_1 + x_2)}{\|x_1 - x_2\|^2}
$$

$$
(x_1 - x_2)^T(x_1 + x_2)\ =\ \|x_1\|^2 - \|x_2\|^2
\ \ \Longrightarrow\ \ 
b \ =\ -\,\frac{\|x_1\|^2 - \|x_2\|^2}{\|x_1 - x_2\|^2}
$$


## ۴) نتیجهٔ هندسی

- بردار نرمالِ ابرصفحه (\(w\)) دقیقاً در راستای \(x_1 - x_2\) است.
- ابرصفحهٔ تصمیم \(w^T x + b = 0\) عمود بر خطِ واصلِ دو نقطه بوده و «در میانهٔ آن‌ها» نسبت به این راستا قرار می‌گیرد (فاصلهٔ برابر از هر دو نمونه بر حسبِ حاشیه).
- بنابراین **دو نقطهٔ داده (یکی از هر کلاس) به‌تنهایی مکان ابرصفحهٔ بیشینه‌حاشیه را به‌طور یکتا تعیین می‌کنند**؛ این موضوع مستقل از بُعدِ فضا است، چون فقط به ضرب داخلی‌ها و نُرم‌ها تکیه دارد.
