# حل تمرین ۱.۱۵ کتاب بیشاپ

## ۱. درک فرمول اولیه

ترم مرتبه‌ی $M$ام برای یک چندجمله‌ای در فضای $D$ بُعدی به صورت زیر نوشته می‌شود:

$$
\sum_{i_1=1}^{D} \sum_{i_2=1}^{D} \cdots \sum_{i_M=1}^{D} w_{i_1 i_2 \cdots i_M} x_{i_1} x_{i_2} \cdots x_{i_M}
$$

تعداد کل جمله‌ها برابر با $D^M$ است. اما به دلیل متقارن بودن حاصل‌ضرب متغیرها (مثلاً $x_1x_2 = x_2x_1$)، بسیاری از ضرایب $w$ تکراری هستند و می‌توان آن‌ها را حذف کرد.


## ۲. بازنویسی با ضرایب متقارن

برای حذف ضرایب تکراری، می‌توان فقط جمله‌هایی را در نظر گرفت که در آن‌ها اندیس‌ها به‌صورت مرتب غیرکاهشی باشند:

$$
\sum_{1 \le i_1 \le i_2 \le \cdots \le i_M \le D} w_{i_1 i_2 \cdots i_M} x_{i_1} x_{i_2} \cdots x_{i_M}
$$

در این حالت، تنها ضرایب مستقل باقی می‌مانند.


## ۳. رابطه‌ بازگشتی

تعداد ضرایب مستقل را با $n(D, M)$ نمایش می‌دهیم. این تعداد با رابطه‌ی زیر به‌صورت بازگشتی تعریف می‌شود:

$$
n(D, M) = \sum_{i=1}^{D} n(i, M-1)
$$

این رابطه بیان می‌کند که برای ساخت یک جمله از مرتبه‌ی $M$ در $D$ بُعد، می‌توان متغیر اول را انتخاب کرد و باقی‌مانده را با استفاده از متغیرهای محدودتر (تا $i$) و مرتبه‌ی $M-1$ ساخت.


## ۴. اثبات فرم بسته با استفاده از استقرا

هدف، اثبات فرمول بسته‌ زیر برای $n(D, M)$ است:

$$
n(D, M) = \frac{(D + M - 1)!}{(D - 1)! M!}
$$

این همان فرمول ترکیب با تکرار (stars and bars) در ترکیبیات است.


## ۵. پایه‌ی استقرا: $D = 1$

در فضای یک‌بعدی فقط یک متغیر $x_1$ وجود دارد، بنابراین فقط یک جمله‌ی $x_1^M$ می‌توان نوشت. پس

$$
n(1, M) = 1
$$

و این با فرمول کلی نیز سازگار است:

$$
\frac{(1 + M - 1)!}{(1 - 1)! M!} = \frac{M!}{0! M!} = 1 \quad \checkmark
$$


 ## ۶. گام استقرا: از ( $ D $ به $ D+1 $)
فرض می‌کنیم فرمول برای $D$ صحیح باشد:

$$
n(D, M) = \frac{(D + M - 1)!}{(D - 1)! M!}
$$

 $(D+1)$ اکنون رابطه‌ بازگشتی برای  

را بررسی می‌کنیم

$$
n(D+1, M) = \sum_{i=1}^{D+1} n(i, M-1)
$$

با استفاده از فرض استقرا:

$$
n(D+1, M) = \frac{1}{(M-1)!} \sum_{i=1}^{D+1} \frac{(i + M - 2)!}{(i - 1)!}
$$

و این مجموع طبق رابطه‌ شناخت ‌شده برابر است با:

$$
\frac{(D + M)!}{D! M!} = n(D+1, M)
$$


## ۷. بررسی حالت خاص: $ M = 2 $

برای مرتبه‌ی دوم، تعداد ضرایب مستقل برابر است با:

$$
n(D, 2) = \frac{(D + 1)!}{(D - 1)! \cdot 2!} = \frac{D(D+1)}{2}
$$

که همان تعداد جمله‌های متقارن از نوع $x_i x_j$ با شرط $i \le j$ است. بنابراین فرمول کلی در این حالت نیز برقرار است.


## نتیجه‌گیری

با استفاده از رابطه‌ی بازگشتی و روش استقرا نشان دادیم که تعداد ضرایب مستقل برای چندجمله‌ای مرتبه‌ی $M$ در فضای $D$ بُعدی برابر است با:

$$
n(D, M) = \frac{(D + M - 1)!}{(D - 1)! M!}
$$

این نتیجه با فرمول ترکیب با تکرار در ریاضیات تطابق کامل دارد.
