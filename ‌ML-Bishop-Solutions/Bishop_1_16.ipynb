# حل تمرین ۱.۱۶ کتاب بیشاپ

## ۱. تعریف تعداد ضرایب مستقل کل

ما در تمرین قبل تعداد ضرایب مستقل در یک ترم با درجه‌ی مشخص $(m)$ را با $( n(D, m) )$ نشان دادیم. حالا تعداد ضرایب مستقل در تمام جمله‌ها تا درجه‌ی $( M )$ برابر است با:

$$
N(D, M) = \sum_{m=0}^{M} n(D, m)
$$

که این همون رابطه‌ (1.138) است.


## ۲. اثبات رابطه‌ (1.138) با استقرا

### پایه استقرا:
برای \( M = 1 \):

$$
N(D, 1) = n(D, 0) + n(D, 1)
$$

که دقیقا با تعریف سازگار است.

### گام استقرا:
فرض میکنیم رابطه برای \( M \) برقرار است. آنگاه:

$$
N(D, M + 1) = N(D, M) + n(D, M + 1)
$$

طبق فرض استقرا داریم:

$$
N(D, M) = \sum_{m=0}^{M} n(D, m) \Rightarrow N(D, M + 1) = \sum_{m=0}^{M+1} n(D, m)
$$

پس رابطه برای \( M+1 \) هم برقرار است.


## ۳. اثبات رابطه‌ی بسته (1.139)

از تمرین قبل می‌دانیم:

$$
n(D, m) = \frac{(D + m - 1)!}{(D - 1)! m!}
$$

با جایگذاری در مجموع:

$$
N(D, M) = \sum_{m=0}^{M} \frac{(D + m - 1)!}{(D - 1)! m!}
$$

این مجموع برابر است با:

$$
N(D, M) = \frac{(D + M)!}{D! M!}
$$

که همان رابطه‌ی (1.139) است و می‌توان با استقرا نیز آن را اثبات کرد.


## ۴. تحلیل نرخ رشد با استفاده از تقریب استرلینگ

تقریب استرلینگ:

$$
n! \approx n^n e^{-n}
$$

### وقتی که$( D \gg M )$:

در این حالت:

$$
N(D, M) = \frac{(D + M)!}{D! M!} \approx D^M
$$

### وقتی که $( M \gg D )$:

با استفاده از تقارن بین D و M در فرمول:

$$
N(D, M) \approx M^D
$$

پس رشد تابع به صورت نمایی با بزرگ‌ترین مقدار بین D و M است.


## ۵. محاسبه‌ عددی

### حالت اول: \( D = 10, M = 3 \)

$$
N(10, 3) = \frac{13!}{10! \cdot 3!} = 286
$$

### حالت دوم: \( D = 100, M = 3 \)

$$
N(100, 3) = \frac{103!}{100! \cdot 3!} = \frac{103 \cdot 102 \cdot 101}{6} = 176851
$$
