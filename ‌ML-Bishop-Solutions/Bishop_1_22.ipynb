# حل تمرین ۱.۲۲ کتاب بیشاپ

## صورت سؤال

رابطه‌ی ریسک مورد انتظار به‌صورت زیر است

$$
R(C_j|x) = \sum_k L_{kj} \, p(C_k | x)
$$

نشان دهید که اگر ماتریس زیان به‌صورت زیر باشد

$$
L_{kj} = 1 - I_{kj}
$$

آنگاه تصمیم ‌گیری بهینه معادل است با انتخاب کلاسی که $p(C_j | x)$ آن بیشترین مقدار را دارد


## تعریف ماتریس زیان خاص

فرض می‌کنیم:

$$
L_{kj} =
\begin{cases}
0 & \text{if } k = j \\
1 & \text{if } k \ne j
\end{cases}
$$

یعنی اگر پیش‌ بینی درست باشد، زیانی نداریم؛ ولی در صورت اشتباه، زیان همیشه برابر ۱ است


## محاسبه ریسک مورد انتظار

با جایگذاری در فرمول داریم

$$
R(C_j|x) = \sum_k (1 - I_{kj}) \, p(C_k|x)
= \sum_k p(C_k|x) - \sum_k I_{kj} \, p(C_k|x)
= 1 - p(C_j|x)
$$

برای کمینه کردن $R(C_j|x)$، باید $p(C_j|x)$ را بیشینه کنیم


## نتیجه ‌گیری

در نتیجه، باید کلاسی را انتخاب کنیم که $p(C_j | x)$ بیشترین مقدار را داشته باشد


## تفسیر ماتریس زیان

این ماتریس زیان نشان می‌دهد که:
- اگر برچسب به‌ درستی پیش‌ بینی شود، زیان صفر است.
- اگر اشتباه شود، زیان برابر با ۱ است، بدون توجه به اینکه اشتباه به کدام کلاس بوده است.

بنابراین این ماتریس فقط به درستی یا نادرستی پیش‌ بینی اهمیت می‌دهد، نه نوع اشتباه.
