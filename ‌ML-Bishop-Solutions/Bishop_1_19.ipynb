# حل نهایی تمرین ۱.۱۹ کتاب بیشاپ

## ۱. تعریف مسئله
یک کره به شعاع $a$ در فضای $D$-بعدی درون یک هایپرکیوب $D$-بعدی با ضلع $2a$ قرار دارد، به‌طوری‌که به مرکز هر وجه مماس است.
 
هدف، محاسبه نسبت حجم کره به حجم هایپرکیوب و بررسی رفتار آن در ابعاد بالا است.

## ۲. حجم کره و مکعب در فضای \( D \)-بعدی

با استفاده از نتیجه تمرین ۱.۱۸:

$$
V_{\text{sphere}} = \frac{2 \pi^{D/2}}{D \cdot \Gamma(D/2)} a^D
$$

و حجم مکعب:

$$
V_{\text{cube}} = (2a)^D = 2^D a^D
$$

در نتیجه، نسبت دو حجم برابر است با:

$$
\frac{V_{\text{sphere}}}{V_{\text{cube}}}
= \frac{2 \pi^{D/2} a^D}{D \cdot \Gamma(D/2) \cdot 2^D a^D}
= \frac{\pi^{D/2}}{D \cdot 2^{D-1} \cdot \Gamma(D/2)} \tag{1.145}
$$


## ۳. میل نسبت حجم‌ها به صفر در ابعاد بالا

برای (  $ D \to \infty $)، از تقریب استرلینگ استفاده می‌کنیم
$$
\Gamma(x + 1) \approx \sqrt{2\pi} \, x^{x + 1/2} e^{-x}
$$

اگر ($ x = D/2 $)، آنگاه

$$
\Gamma(D/2) \sim \left(\frac{D}{2}\right)^{D/2 - 1/2} e^{-D/2}
$$

از آن‌جا که $$( \pi^{D/2} )$$ رشد نمایی دارد، اما $$( 2^{D} \cdot \Gamma(D/2) )$$ رشد سریع‌تری دارد، می‌توان نتیجه گرفت که:

$$
\frac{\pi^{D/2}}{D \cdot 2^{D-1} \cdot \Gamma(D/2)} \to 0 \quad \text{when } D \to \infty
$$


## ۴. نسبت فاصله مرکز تا گوشه و تا وجه

### فاصله از مرکز تا وسط وجه:

فقط یکی از مؤلفه‌ها برابر \( a \) است و بقیه صفر، بنابراین:

$$
\|s\| = a
$$

### فاصله از مرکز تا گوشه:

تمام مؤلفه‌ها برابر ($ \pm a$) هستند:

$$
\|c\| = \sqrt{a^2 + a^2 + \dots + a^2} = a \sqrt{D}
$$

### نسبت این دو فاصله:

$$
\frac{\|c\|}{\|s\|} = \frac{a \sqrt{D}}{a} = \sqrt{D}
$$

و بنابراین:

$$
\lim_{D \to \infty} \sqrt{D} = \infty
$$


## نتیجه‌گیری نهایی

- در فضای با ابعاد بالا، نسبت حجم کره به مکعب به صفر میل می‌کند.
- فاصله گوشه‌ها از مرکز بسیار بزرگ‌تر از فاصله تا وجه می‌شود.
- نتیجه: در ابعاد بالا بیشتر حجم مکعب در گوشه‌ها متمرکز می‌شود، که ساختاری شبیه شاخک‌های بلند دارد.
