# حل تمرین ۱.۲۴ کتاب بیشاپ

## صورت سؤال

در یک مسئله‌ی طبقه‌بندی، اگر زیان نسبت دادن کلاس $C_k$ به کلاس $C_j$ برابر باشد با:

$$
L_{kj}
$$

و اگر گزینه‌ی رد کردن را انتخاب کنیم، زیان آن مقدار ثابت $
\lambda
$ باشد:



می‌خواهیم معیار تصمیم‌گیری بهینه برای کمینه کردن زیان مورد انتظار را بیابیم.

همچنین نشان دهید که در حالت خاصی که:

$$
L_{kj} = 1 - I_{kj}
$$

این معیار معادل با معیار رد کردن کلاسیکی است که در بخش ۱.۵.۳ بیان شده است. ارتباط بین $\lambda$ و آستانه‌ی رد $\theta$ را نیز مشخص کنید.


## ریسک مورد انتظار

اگر بردار ورودی $x$ به کلاس $C_j$ نسبت داده شود، ریسک شرطی به‌صورت زیر تعریف می‌شود:

$$
R(C_j | x) = \sum_k L_{kj} \, p(C_k | x)
$$

اگر تصمیم بگیریم که نمونه را رد کنیم، ریسک آن ثابت و برابر با:

$$
\lambda
$$

خواهد بود.


## معیار تصمیم‌گیری

ابتدا کلاس $j$ را پیدا می‌کنیم که کمترین ریسک شرطی را دارد:

$$
j = \arg\min_l \sum_k L_{kl} \, p(C_k | x)
$$

اگر مقدار ریسک کمینه کمتر از $\lambda$ باشد، آن کلاس را انتخاب می‌کنیم. در غیر این صورت، تصمیم به رد کردن می‌گیریم.


## بررسی حالت خاص

حالتی را در نظر می‌گیریم که ماتریس زیان به‌شکل زیر باشد:

$$
L_{kj} = 1 - I_{kj}
$$

در این صورت، ریسک شرطی برابر خواهد بود با:

$$
R(C_j | x) = 1 - p(C_j | x)
$$

در نتیجه، معیار تصمیم‌گیری به شکل زیر در می‌آید:

$$
\min_j (1 - p(C_j | x)) < \lambda
\Rightarrow
\max_j p(C_j | x) > 1 - \lambda
$$

که این دقیقاً معادل با معیار رد کردن کلاسیکی است اگر داشته باشیم:

$$
\theta = 1 - \lambda
$$


## نتیجه‌گیری

در حضور گزینه‌ی رد کردن:

- اگر کمترین ریسک شرطی از ($\lambda$) کمتر باشد، آن کلاس را انتخاب می‌کنیم.
- اگر نه، نمونه را رد می‌کنیم.

در حالت خاصی که $$L_{kj} = 1 - I_{kj}$$ باشد، این تصمیم‌گیری دقیقاً معادل با معیار رد کلاسیک است به شرطی که $$\theta = 1 - \lambda $$ باشد.
