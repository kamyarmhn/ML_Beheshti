# حل تمرین ۱.۲۳ کتاب بیشاپ

## صورت سؤال

معیار تصمیم‌گیری بهینه را برای کمینه کردن زیان مورد انتظار در حالتی که:
- ماتریس زیان $L_{kj}$ کلی است،
- و احتمال‌های پیشین $p(C_k)$ نیز متفاوت‌اند

بیابید.


## تعریف ریسک مورد انتظار کلی

ریسک مورد انتظار برای فضای تصمیم ‌گیری به‌صورت زیر تعریف می‌شود:

$$
E[L] = \sum_j \int_{R_j} \sum_k L_{kj} \, p(x, C_k) \, dx
$$

با استفاده از قانون ضرب احتمال، می‌ دانیم که:

$$
p(x, C_k) = p(C_k) \, p(x | C_k)
$$

در نتیجه:

$$
E[L] = \sum_j \int_{R_j} \sum_k L_{kj} \, p(C_k) \, p(x | C_k) \, dx
$$


## تعریف ماتریس زیان جدید برای ساده‌سازی

برای ساده ‌تر شدن بیان ریاضی، ماتریس زیان جدید را به‌شکل زیر تعریف می‌کنیم:

$$
L^*_{kj} = L_{kj} \, p(C_k)
$$

و بنابراین:

$$
E[L] = \sum_j \int_{R_j} \sum_k L^*_{kj} \, p(x | C_k) \, dx
$$


 ## تصمیم ‌گیری نقطه‌ ای برای هر ($x$) د

برای هر بردار ورودی $x$، هدف ما کمینه کردن ریسک شرطی است:

$$
R(C_j | x) = \sum_k L_{kj} \, p(C_k | x)
$$

در نتیجه، باید کلاسی را انتخاب کنیم که $R(C_j | x)$ آن کمترین مقدار را داشته باشد.


## نتیجه‌ گیری

در حالت کلی، معیار تصمیم‌گیری برای $x$ به‌شکل زیر است:

- برای هر کلاس $C_j$، ریسک شرطی $R(C_j | x)$ را محاسبه کن.
- کلاسی را انتخاب کن که این مقدار کمینه باشد.

این نتیجه تعمیم‌ یافته‌ از تمرین ۱.۲۲ است، که در آن احتمال‌های پیشین و ماتریس زیان یکسان فرض نشده‌ اند.
