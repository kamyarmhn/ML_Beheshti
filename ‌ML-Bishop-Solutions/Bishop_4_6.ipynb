# حل تمرین 4.6: ارتباط بین روش حداقل مربعات و تحلیل فیشر


**هدف مسئله:**  
در این تمرین می‌خواهیم نشان دهیم که جوابی که از روش **حداقل مربعات** (Least Squares) برای طبقه‌بندی به دست می‌آید، ارتباط مستقیمی با مفاهیم اصلی **تحلیل تشخیص خطی فیشر** (Fisher's Discriminant)، یعنی ماتریس‌های کوواریانس درون‌کلاسی ($S_W$) و بین‌کلاسی ($S_B$) دارد.

---

## شروع از فرمول اصلی حداقل مربعات
$$
W = (\Phi^T \Phi)^{-1} \Phi^T T
$$

یا به شکل دیگر:
$$
(\Phi^T \Phi) W = \Phi^T T
$$



## گام اول: محاسبه $\Phi^T \Phi$
ماتریس $\Phi$ (ماتریس طراحی) هر سطر آن، یک بردار ویژگی $\phi(x_n)^T$ است. بنابراین:
$$
\Phi^T \Phi = \sum_{n=1}^N \phi_n \phi_n^T
$$

با گروه‌بندی بر اساس کلاس‌ها:
$$
\Phi^T \Phi = \sum_{k=1}^K \sum_{n \in C_k} \phi_n \phi_n^T
$$

ماتریس کوواریانس درون‌کلاسی:
$$
S_W = \sum_{k=1}^K S_k, \quad S_k = \sum_{n \in C_k} (\phi_n - m_k)(\phi_n - m_k)^T
$$

با بسط $S_k$ و استفاده از $\sum_{n \in C_k} \phi_n = N_k m_k$ داریم:
$$
\sum_{n \in C_k} \phi_n \phi_n^T = S_k + N_k m_k m_k^T
$$

در نتیجه:
$$
\Phi^T \Phi = S_W + \sum_{k=1}^K N_k m_k m_k^T
$$



## گام دوم: محاسبه $\Phi^T T$
ماتریس $T$ یک ماتریس $N \times K$ است با کدگذاری 1-of-K. ستون $k$ام $\Phi^T T$ برابر است با:
$$
\sum_{n \in C_k} \phi_n = N_k m_k
$$



## گام سوم: ترکیب نتایج
با جایگذاری نتایج گام اول و دوم:
$$
\big( S_W + \sum_{k=1}^K N_k m_k m_k^T \big) W = (N_1 m_1, N_2 m_2, \dots, N_K m_K)
$$

---

**نتیجه:**  
این معادله نشان می‌دهد که $W$ به طور مستقیم به $S_W$ و $m_k$ وابسته است، که همان مفاهیم کلیدی تحلیل فیشر هستند.
